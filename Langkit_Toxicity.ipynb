```python
import pandas as pd
from langkit import toxicity, extract
```

```python
df = pd.read_csv("Ethos_Dataset_Binary.csv", delimiter=';')
```

```python
toxicity.init(model_path="martin-ha/toxic-comment-model")
```

```python
df['toxicity']=""
```

```python
for i, row in df.iterrows():
    # Apply extract function to each tweet and store result in toxicity column
    df.at[i, 'toxicity'] = extract({"prompt": row['comment']})
```

```python
def extract_toxicity_score(toxicity_data):
    # Check if 'prompt.toxicity' exists in the dictionary
    return toxicity_data.get('prompt.toxicity', None)
```

```python
df['toxicity_score_martin'] = df['toxicity'].apply(extract_toxicity_score)
```
